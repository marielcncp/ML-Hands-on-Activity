# Round 1: Modeling without Augmentation

```python
from pathlib import Path

# Project root (since notebook is inside ML-Hands-on-Activity)
PROJECT_DIR = Path.cwd()

TRAIN_DIR = PROJECT_DIR / "LT01_digits"
TEST_DIR  = PROJECT_DIR / "LT10_digits"

# Safety checks
assert TRAIN_DIR.exists(), f"Training folder not found: {TRAIN_DIR}"
assert TEST_DIR.exists(),  f"Test folder not found: {TEST_DIR}"

print("TRAIN_DIR:", TRAIN_DIR)
print("TEST_DIR :", TEST_DIR)
```

```python
# ============================================================
# PHASE 1 — Frozen Backbone
# ============================================================

model = build_resnet18(num_classes=10, freeze_backbone=True).to(device)

model, hist_frozen = train(
    model,
    train_dl,
    val_dl,
    epochs=10,
    lr=1e-3
)

# ============================================================
# PHASE 2 — Fine-Tuning layer4 + fc
# ============================================================

for name, p in model.named_parameters():
    if name.startswith("layer4") or name.startswith("fc"):
        p.requires_grad = True

model, hist_ft = train(
    model,
    train_dl,
    val_dl,
    epochs=5,
    lr=1e-4
)

# ============================================================
# FINAL EVALUATION ON LT10 (TRUE TEST)
# ============================================================

acc, report, cm = evaluate(model, TEST_DIR)

print(f"\nTEST Accuracy (LT10 exchanged): {acc:.4f}\n")
print("Classification Report (Test = LT10):")
print(report)

print("Confusion Matrix (Test = LT10):")
print(cm)
```

```python
# ============================================================
# PLOT 1 — Validation Accuracy Curve
# ============================================================

v1 = hist_frozen["val_acc"]
v2 = hist_ft["val_acc"]

x1 = list(range(1, len(v1) + 1))
x2 = list(range(len(v1) + 1, len(v1) + len(v2) + 1))

plt.figure(figsize=(8,5))

plt.plot(x1, v1, marker="o", label="Val Acc — Frozen")
plt.plot(x2, v2, marker="o", label="Val Acc — Fine-tune")

all_val = v1 + v2
best_val = max(all_val)
final_val = all_val[-1]

metrics_text = (
    f"Best Val Acc (LT01): {best_val:.4f}\n"
    f"Final Val Acc (LT01): {final_val:.4f}\n"
    f"Test Acc (LT10): {acc:.4f}"
)

plt.gca().text(
    0.02, 0.98, metrics_text,
    transform=plt.gca().transAxes,
    va="top",
    ha="left",
    bbox=dict(boxstyle="round", facecolor="white", alpha=0.85)
)

plt.title("Validation Accuracy per Epoch\n(Train/Val on LT01; Test on LT10)")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.ylim(0, 1.0)
plt.legend()
plt.show()
```

```python
# ============================================================
# PLOT 2 — Confusion Matrix Heatmap
# ============================================================

plt.figure(figsize=(7,6))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=range(10),
    yticklabels=range(10)
)

plt.title("Confusion Matrix (Test = LT10)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

# Round 2: Modeling with Data Augmentaiton

## Preprocessing

```python
# ============================================================
# Use geometric transformations, color augmentation, and normalization 
# ============================================================

from torchvision import transforms
from pretrained_pipeline import DigitsDataset

imagenet_mean = [0.485, 0.456, 0.406]
imagenet_std  = [0.229, 0.224, 0.225]

# Stronger augmentation
train_tfms_aug = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomRotation(20),
    transforms.RandomAffine(
        degrees=0,
        translate=(0.15, 0.15),
        scale=(0.85, 1.15),
        shear=(-10, 10),
    ),
    transforms.ColorJitter(
        brightness=0.3,
        contrast=0.3
    ),
    transforms.RandomPerspective(distortion_scale=0.2, p=0.5),
    transforms.ToTensor(),
    transforms.Normalize(imagenet_mean, imagenet_std),
])

val_tfms_aug = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(imagenet_mean, imagenet_std),
])
```

```python
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split
import torch

# Stratified split again (same seed for fairness)
labels = [y for _, y in lt01_items]

train_items_aug, val_items_aug = train_test_split(
    lt01_items,
    test_size=0.2,
    stratify=labels,
    random_state=SEED
)

train_ds_aug = DigitsDataset(train_items_aug, train_tfms_aug)
val_ds_aug   = DigitsDataset(val_items_aug, val_tfms_aug)

g = torch.Generator()
g.manual_seed(SEED)

train_dl_aug = DataLoader(
    train_ds_aug,
    batch_size=32,
    shuffle=True,
    generator=g
)

val_dl_aug = DataLoader(
    val_ds_aug,
    batch_size=32,
    shuffle=False
)
```

## Modeling fr this time

```python
model_aug = build_resnet18(num_classes=10, freeze_backbone=True).to(device)

# Frozen Backbone
model_aug, hist_frozen_aug = train(
    model_aug,
    train_dl_aug,
    val_dl_aug,
    epochs=10,
    lr=1e-3
)

# Fine-tune layer4 + fc
for name, p in model_aug.named_parameters():
    if name.startswith("layer4") or name.startswith("fc"):
        p.requires_grad = True

model_aug, hist_ft_aug = train(
    model_aug,
    train_dl_aug,
    val_dl_aug,
    epochs=5,
    lr=1e-4
)

# Evaluate
acc_aug, report_aug, cm_aug = evaluate(model_aug, TEST_DIR)

print(f"\nTEST Accuracy with Augmentation (LT10): {acc_aug:.4f}\n")
print(report_aug)
print(cm_aug)
```

## Graph

```python
# Combine histories
val_round1 = hist_frozen["val_acc"] + hist_ft["val_acc"]
val_round2 = hist_frozen_aug["val_acc"] + hist_ft_aug["val_acc"]

epochs1 = range(1, len(val_round1) + 1)
epochs2 = range(1, len(val_round2) + 1)

plt.figure(figsize=(8,5))

plt.plot(epochs1, val_round1, marker="o", label="Round 1 — No Aug")
plt.plot(epochs2, val_round2, marker="o", label="Round 2 — With Aug")

plt.title("Validation Accuracy Comparison (LT01)")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.ylim(0, 1.0)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.heatmap(
    cm_aug,
    annot=True,
    fmt="d",
    cmap="Greens",
    xticklabels=range(10),
    yticklabels=range(10)
)

plt.title("Confusion Matrix (LT10) — With Augmentation")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```
